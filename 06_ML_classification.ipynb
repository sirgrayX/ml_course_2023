```python
from sklearn.base import BaseEstimator

class Never5Classifier(BaseEstimator):
    def fit(self, X, y=None):
        pass
    
    def predict(self, X):
        return np.zeros((len(X), 1), dtype=bool)
    
never_5_clf = Never5Classifier()
cross_val_score(never_5_clf, X_train, y_train_5, cv=3,
               scoring='accuracy')
```

```python
from sklearn.metrics import precision_recall_curve
y_scores = cross_val_predict(sgd_clf, X_train, y_train_5, cv=3,
                            method='decision_function')
#   ,         
precisions, recalls, thresholds = precision_recall_curve(y_train_5, y_scores)
#   ,    
plt.figure(figsize=(8, 5))
plt.plot(thresholds, precisions[:-1], 'b--', label='')
plt.plot(thresholds, recalls[:-1], 'r-', label='')
plt.grid(True)
plt.xlabel('')
plt.xlim(-thresholds.max(), thresholds.max())
plt.legend()
plt.show();
```

 roc_curve()    ,         .       :

```python
def plot_roc_curve(fpr, tpr, label=None):
    plt.plot(fpr, tpr, linewidth=2, label=label)
    plt.plot([0, 1], [0, 1], 'k--') #  
    plt.xlabel('  ')
    plt.ylabel('   ')
    plt.grid(True)
```

```python
y_scores_forest = y_probs_forest[:, 1]
fpr_forest, tpr_forest, \
      thresholds_forest = roc_curve(y_train_5, y_scores_forest)
#    roc-
plt.figure(figsize=(8, 5))
plt.plot(fpr, tpr, 'b:', label='SGD')
plot_roc_curve(fpr_forest, tpr_forest, ' ')
plt.legend(loc='lower right')
plt.show();
```
