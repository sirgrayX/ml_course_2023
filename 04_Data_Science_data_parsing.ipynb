````python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

html = requests.get('https://sunlight.net/catalog').text  #  html  
soup = bs(html)  #    BeautifulSoup

data_dict = {
    'url' : [],
    'name' : [],
    'price' : []
}

links = soup.find_all('a', class_='cl-item-link js-cl-item-link js-cl-item-root-link')  #     
for i, link in enumerate(links):
    url = link.get("href")  #   
    name = link.get_text()  #      
    price = soup.find_all("div", class_="cl-item-info-price-discount")[i].get_text()  #  
    print(f" {i}\nURL: {url}\nName: {name}\nPrice: {price}\n\n")
#     print(i)
#     print(f"Url - {url}")
#     print(f"Name - {name}")
#     print(f"Price - {price}\n")
    data_dict['url'].append(url)
    data_dict['name'].append(name)
    data_dict['price'].append(price)
````
